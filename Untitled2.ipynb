```python
import requests
from bs4 import BeautifulSoup
```

```python
html = requests.get('http://www.pngmart.com/image/category/art')
soup = BeautifulSoup(html.text)
```

```python
img_url = soup.select('img.wp-post-image')[0]['src'][:-12]+'.png'
```

```python
img_data = requests.get(img_url)
```

```python
dir(img_data)
img_data.text
```

```python
file_name = img_url.split('/')[-1]
```

```python
with open(file_name,'bw') as f:
    f.write(img_data.raw)
```
