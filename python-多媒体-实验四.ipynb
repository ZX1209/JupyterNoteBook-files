1．编写绘制曲线e - 0 .7 t cos (5 t)的函数, 其中t 为[ 0 , 10 ] , 增量为0 .01。
2．为估算程序的运行时间,Matlab 内置有用于开始计时的命令( tic) 和终止计时的命令

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(0,10+0.01,0.01)
```

```python
y = np.e-(0.7*t)*np.cos(5*t)
```

```python
plt.plot(t,y);plt.show()
```

# 使用 %%timeit 魔术方法 测试运行时间

```python
%%timeit
t = np.arange(0,10+0.01,0.01)
y = np.e-(0.7*t)*np.cos(5*t)
plt.plot(t,y);plt.show()
```
