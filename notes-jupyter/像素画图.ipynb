```python
from PIL import Image, ImageDraw
im = Image.new('RGBA',(1000,1000),'white')
draw =  ImageDraw.Draw(im)
```

```python
point_list = []

start_x,start_y=(500,500)
end_x,end_y=(839,207)

dx=start_x-end_x
dy=start_y-end_y

k = dy/dx



if abs(dx)>abs(dy):
    y=start_y
    for x in range(start_x,end_x+1):
        point_list.append((x,round(y)))
        y+=k;
        
```

```python
print(point_list)
```

```python
draw.point(point_list,'red')

draw.line([(400,400),(739,107)],'blue')
im.save('.\imge_draw_test.png')
```
