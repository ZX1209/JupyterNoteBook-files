```python
N=500;
M=3;

def remove0(x):
    if x==0:
        return False
    else:
        return True
    

people = list(range(1,N+1));
i=0;
while N>M:
    global i;
    if (i+1)%M!=0: people[i]=0;
    i+=1;
    
    if(i>=N):
        i=1;
        people = list(filter(remove0,people))
        N=len(people)
        
print(people)
    
    
```
