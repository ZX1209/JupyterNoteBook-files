## 随机数

```python
import random
random.choice(['apple','peer','bannana']) #随机选择

random.sample(range(100),10) #取样

random.random() #随机浮点数

random.randrange(6) #从range(6)中去一个随机数
```

## 互联网访问
urllib.request
smtplib

```python
from urllib.request import urlopen
for line in urlopen('http://tycho.usno.navy.mil/cgi-bin/timer.pl'):
    line = line.decode('utf-8')  # Decoding the binary data to text.
    if 'EST' in line or 'EDT' in line:
        print(line)
```
